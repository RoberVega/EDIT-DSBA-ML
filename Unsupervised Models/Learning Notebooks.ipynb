# Unsupervised Models

Unsupervised models are those that we can construct whenever we do not have a target feature to fit. These are based on finding similarities or relationships between the data in such a way that we can organise them into groups.

**Clustering** is a set of techniques used to partition data into groups, or clusters. Clusters are loosely defined as groups of data objects that are more similar to other objects in their cluster than they are to data objects in other clusters. We will discuss the main three types of models:

</br>
 
1. **Partitional clustering** divides data objects into nonoverlapping groups. In other words, no object can be a member of more than one cluster, and every cluster must have at least one object. 

    Partitional clustering methods have several strengths:
     - They work well when clusters have a spherical shape.
     - They’re scalable with respect to algorithm complexity.

    They also have several weaknesses:
     - They’re not well suited for clusters with complex shapes and different sizes.
     - They break down when used with clusters of different densities.
    
 </br>   
    

2. **Hierarchical clustering** determines cluster assignments by building a hierarchy. This is implemented by either a bottom-up or a top-down approach:

     - **Agglomerative clustering** is the bottom-up approach. It merges the two points that are the most similar until all points have been merged into a single cluster.
     - **Divisive clustering** is the top-down approach. It starts with all points as one cluster and splits the least similar clusters at each step until only single data points remain.
     
     The strengths of hierarchical clustering methods include the following:

    - They often reveal the finer details about the relationships between data objects.
    - They provide an interpretable dendrogram.
 
    The weaknesses of hierarchical clustering methods include the following:

    - They’re computationally expensive with respect to algorithm complexity.
    - They’re sensitive to noise and outliers.
    
</br>

3. **Density-based clustering** determines cluster assignments based on the density of data points in a region. Clusters are assigned where there are high densities of data points separated by low-density regions.

    This approach doesn’t require the user to specify the number of clusters. Instead, there is a distance-based parameter that acts as a tunable threshold. This threshold determines how close points must be to be considered a cluster member.
    
    The strengths of density-based clustering methods include the following:

    - They excel at identifying clusters of nonspherical shapes.
    - They’re resistant to outliers.

    The weaknesses of density-based clustering methods include the following:

    - They aren’t well suited for clustering in high-dimensional spaces.
    - They have trouble identifying clusters of varying densities.


## $k$-means

The $k$-means clustering method is an unsupervised machine learning technique used to identify clusters of data objects in a dataset. Conventional $k$-means requires only a few steps. The first step is to randomly select $k$ centroids, where $k$ is equal to the number of clusters you choose. Centroids are data points representing the center of a cluster.

The main element of the algorithm works by a two-step process called expectation-maximization:

1. The **expectation** step assigns each data point to its nearest centroid. 
2. The **maximization** step computes the mean of all the points for each cluster and sets the new centroid. 




```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

features, true_labels = make_blobs(
    n_samples=1000,
    centers=3,
    cluster_std=2.75,
    random_state=42
)


scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)


plt.scatter(scaled_features[:,0],scaled_features[:,1])
plt.title('Data')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)

kmeans.fit(scaled_features)

labels=kmeans.labels_

colors = ['y', 'g', 'b']

# Plot the data points with different colors for each cluster
for i in range(len(colors)):
    plt.scatter(scaled_features[labels == i, 0], scaled_features[labels == i, 1], c=colors[i], label='Cluster {}'.format(i))

# Add the cluster centers
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c='black', s=50)
plt.legend()
plt.title('$k$-means')

```

## $k$-means++

This method is the same as $k$-means, but with an algorithm for choosing the initial values of the centroids that reduces the complexity of the problem. The algorithm works in the following steps:

1. Choose one center uniformly at random among the data points.
2. For each data point $x$ not chosen yet, compute $D(x)$, the distance between $x$ and the nearest center that has already been chosen.
3. Choose one new data point at random as a new center, using a weighted probability distribution where a point $x$ is chosen with probability proportional to $D(x)^2$.
4. Repeat Steps 2 and 3 until k centers have been chosen.

```python
from sklearn.cluster import KMeans

kmeanspp = KMeans(
    init="k-means++",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)

kmeanspp.fit(scaled_features)

labels=kmeanspp.labels_

colors = ['y', 'g', 'b']

# Plot the data points with different colors for each cluster
for i in range(len(colors)):
    plt.scatter(scaled_features[labels == i, 0], scaled_features[labels == i, 1], c=colors[i], label='Cluster {}'.format(i))

# Add the cluster centers
plt.scatter(kmeanspp.cluster_centers_[:,0],kmeanspp.cluster_centers_[:,1], c='black', s=50)
plt.legend()
plt.title('$k$-means++')

```

## Fuzzy $c$-means

In the Fuzzy c-means (FCM) clustering method, we have two parameters, $μ_{ij}$ and $c_i$, and one hyperparameter, $m$.

- $μ_{ij}$, the membership value, is the probability that the $j$-th data point belongs to the $i$-th cluster. 
- $c_i$ is the center of the $i$-th cluster
- $m$ is the fuzzifier, which controls how fuzzy the cluster boundary should be.

```python
#!pip install fuzzy-c-means
import numpy as np
from fcmeans import FCM

fuzzymeans = FCM(n_clusters=3) # we use two cluster as an example
fuzzymeans.fit(scaled_features)
fuzzymeans.centers

probs = fuzzymeans.soft_predict(scaled_features)

# create a scatter plot with color-coded points
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=probs)

# Add the cluster centers
plt.scatter(fuzzymeans.centers[:,0],fuzzymeans.centers[:,1], c='black', s=50)
plt.title('Fuzzy $c$-means')
```

## Density-Based Spatial Clustering of Applications with Noise (DBSCAN)

DBSCAN is a base algorithm for density-based clustering. It can discover clusters of different shapes and sizes from a large amount of data, which is containing noise and outliers.

The DBSCAN algorithm uses two parameters:

- minPts: The minimum number of points (a threshold) clustered together for a region to be considered dense.
- eps (ε): A distance measure that will be used to locate the points in the neighborhood of any point. 

These parameters can be understood if we explore two concepts called Density Reachability and Density Connectivity.

**Reachability** in terms of density establishes a point to be reachable from another if it lies within a particular distance (eps) from it.

**Connectivity**, on the other hand, involves a transitivity based chaining-approach to determine whether points are located in a particular cluster.

There are three types of points after the DBSCAN clustering is complete:


- **Core** — This is a point that has at least m points within distance n from itself.
- **Border** — This is a point that has at least one Core point at a distance n.
- **Noise** — This is a point that is neither a Core nor a Border. And it has less than m points within distance n from itself. 


The algorithm proceeds by arbitrarily picking up a point in the dataset (until all points have been visited). If there are at least ‘minPoint’ points within a radius of ‘ε’ to the point then we consider all these points to be part of the same cluster. The clusters are then expanded by recursively repeating the neighborhood calculation for each neighboring point.

There is no way to choose these parameters, but there are rules of thumb to have an initial proposal.


- minPts: a minimum minPts can be derived from the number of dimensions $D$ in the data set, as minPts ≥ D + 1. The minimum value that makes sense is 3, while larger values are usually better for datasets with noise and will yield more significant clusters. As a rule of thumb, minPts $= 2\cdot D$ can be used, but it may be necessary to choose larger values for very large data, for noisy data or for data that contains many duplicates.
- ε: The value for ε can then be chosen by using a $k$-distance graph, plotting the distance to the $k =$ minPts$-1$ nearest neighbor ordered from the largest to the smallest value. Good values of ε are where this plot shows an “elbow”: if ε is chosen much too small, a large part of the data will not be clustered; whereas for a too high value of ε, clusters will merge and the majority of objects will be in the same cluster. In general, small values of ε are preferable, and as a rule of thumb, only a small fraction of points should be within this distance of each other.
    

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler


features, true_labels = make_moons(
    n_samples=250,
    noise=0.05,
    random_state=42
)

scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import DBSCAN, KMeans

kmeans = KMeans(n_clusters=2)
dbscan = DBSCAN(eps=0.3)

# Fit the algorithms to the features
kmeans.fit(scaled_features)
dbscan.fit(scaled_features)


klabels=kmeans.labels_
dbslabels=dbscan.labels_
```

```python
# Plot the data points with different colors for each cluster
colors = ['y', 'g']

for i in range(len(colors)):
    plt.scatter(scaled_features[klabels == i, 0], scaled_features[klabels == i, 1], c=colors[i], label='Cluster {}'.format(i))

# Add the cluster centers
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c='black', s=50)
plt.title('$k$-means')

plt.legend()
```

```python
# Plot the data points with different colors for each cluster
colors = ['y', 'g']

for i in range(len(colors)):
    plt.scatter(scaled_features[dbslabels == i, 0], scaled_features[dbslabels == i, 1], c=colors[i], label='Cluster {}'.format(i))
plt.title('DBSCAN')
plt.legend()
```

## $k$-Medoids

The $k$-medoids method is a clustering algorithm that aims to partition a dataset into $k$ clusters, where each cluster is represented by a single data point called a medoid. The medoid is the data point that minimizes the sum of dissimilarities between it and all the other data points in the cluster.

The $k$-medoids algorithm starts by randomly selecting $k$ medoids from the dataset. Then, it assigns each data point to the nearest medoid based on a chosen distance metric, such as Euclidean distance or Manhattan distance. After that, it computes the sum of dissimilarities between each data point and its assigned medoid.

Next, the algorithm tries to improve the clustering by swapping one of the medoids with a non-medoid data point and recalculating the sum of dissimilarities for each cluster. If the sum of dissimilarities decreases, the swap is accepted, and the algorithm moves on to the next iteration. Otherwise, the swap is rejected, and the algorithm moves on to the next medoid.

The $k$-medoids algorithm iterates until either the medoids no longer change or a maximum number of iterations is reached. At the end of the algorithm, the medoids represent the final clusters.

Compared to the $k$-means algorithm, which uses the mean of each cluster to represent the cluster center, the $k$-medoids algorithm is more robust to outliers since it uses a medoid, which is a real data point, as the cluster center. However, the $k$-medoids algorithm is generally slower than $k$-means since it involves more computations to find the best medoid for each cluster.

```python
#!pip install scikit-learn-extra
from sklearn_extra.cluster import KMedoids

kmedoids = KMedoids(n_clusters=2)

kmedoids.fit(scaled_features)

kmlabels=kmedoids.labels_

colors = ['y', 'g']

for i in range(len(colors)):
    plt.scatter(scaled_features[kmlabels == i, 0], scaled_features[kmlabels == i, 1], c=colors[i], label='Cluster {}'.format(i))

# Add the cluster centers
plt.scatter(kmedoids.cluster_centers_[:,0],kmedoids.cluster_centers_[:,1], c='black', s=50)
plt.legend()
plt.title('$k$-medoids')
```

## Hierarchical Clustering

Hierarchical clustering is an unsupervised learning method for clustering data points. The algorithm builds clusters by measuring the dissimilarities between data. This method can be used on any data to visualize and interpret the relationship between individual data points.

There are two types of hierarchical clustering: agglomerative and divisive.


### Agglomerative clustering

1. We start by assign each point to an individual cluster.

2. Then, at each iteration, we merge the closest pair of clusters and repeat this step until only a the end number of clusters is left.

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(scaled_features)

plt.scatter(scaled_features[:,0], scaled_features[:,1], c=labels)
plt.title('Agglomerative clustering')

plt.show() 
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linkage_data = linkage(scaled_features, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show() 
```

### Divisive clustering

Divisive hierarchical clustering works in the opposite way. 

1. We start with a single cluster and assign all the points to that cluster.

2. Now, at each iteration, we split the farthest point in the cluster and repeat this process until each cluster only contains a single point.

## Biclustering

Biclustering algorithms simultaneously cluster rows and columns of a data matrix. These clusters of rows and columns are known as biclusters. Each determines a submatrix of the original data matrix with some desired properties.

```python
from sklearn.cluster import SpectralBiclustering

bicluster = SpectralBiclustering(n_clusters=1)
bicluster.fit(scaled_features)

fit_data = scaled_features[np.argsort(bicluster.row_labels_)]
fit_data = fit_data[:, np.argsort(bicluster.column_labels_)][:10,:]

plt.matshow(fit_data, cmap=plt.cm.Blues)

#plt.scatter(scaled_features[:,0], scaled_features[:,1], c=labels)
plt.show() 
```

The following is a more telling example, but using a cooked dataset.

```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.datasets import make_checkerboard
from sklearn.cluster import SpectralBiclustering

n_clusters = (4, 3)
data, rows, columns = make_checkerboard(
    shape=(300, 300), n_clusters=n_clusters, noise=10, shuffle=False, random_state=0
)

# shuffle clusters
rng = np.random.RandomState(0)
row_idx = rng.permutation(data.shape[0])
col_idx = rng.permutation(data.shape[1])
data = data[row_idx][:, col_idx]

plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Shuffled dataset")

model = SpectralBiclustering(n_clusters=n_clusters, method="log", random_state=0)
model.fit(data)

fit_data = data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]

plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; rearranged to show biclusters")

plt.show()
```
